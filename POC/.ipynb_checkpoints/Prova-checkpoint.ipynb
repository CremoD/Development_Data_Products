```python
forecasts_rows = []

for elem in response_json:
    curr_row = []
    
    dt_iso = elem['dt_txt']
    temp = round(elem['main']['temp']-273.15, 2)
    feels_like = round(elem['main']['feels_like']-273.15, 2)
    temp_min = round(elem['main']['temp_min']-273.15, 2)
    temp_max = round(elem['main']['temp']-273.15, 2)
    hum = elem['main']['humidity']
    wind = elem['wind']['speed']
    weather = elem['weather'][0]['main']
    curr_row.append(dt_iso)
    #curr_row.append(station)
    curr_row.append(temp)
    curr_row.append(feels_like)
    curr_row.append(temp_min)
    curr_row.append(temp_max)
    curr_row.append(hum)
    curr_row.append(wind)
    curr_row.append(weather)
    
    #iterate over stations
    for station in stations:
        temp_row = curr_row[:]
        temp_row.insert(1, station)
        #append current row
        forecasts_rows.append(temp_row)
    
```

# Apply tranformations to new dataset

```python
def extract_from_date(df):
    
    # extract year
    df['year'] = df['start_date'].dt.year
    # extract month
    df['month'] = df['start_date'].dt.month
    # extract day
    df['day'] = df['start_date'].dt.day
    # extract day_of_week (0 monday - 6 sunday)
    df['day_of_week']=df['start_date'].dt.dayofweek
    # extract hour
    df['hour']= df['start_date'].dt.hour
    
    # extract workday boolean
    df['day_type'] = np.where(df['day_of_week']<5, 'workday', 'weekend')
    
    # extract season
    conditions_weather = [
        df['month'].isin([1,2,12]),
        df['month'].isin([3,4,5]),
        df['month'].isin([6,7,8]),
        df['month'].isin([9,10,11])
    ]
    choices = ['winter', 'spring', 'summer', 'fall']
    df['season'] = np.select(conditions_weather, choices)
    
    # extract day period
    conditions_period = [
        df['hour'].isin([0,1,2,3,4,5]),
        df['hour'].isin([6,7,8,9,10,11]),
        df['hour'].isin([12,13,14,15,16,17]),
        df['hour'].isin([18,19,20,21,22,23])
    ]
    choices = ['night', 'morning', 'afternoon', 'evening']
    df['day_period'] = np.select(conditions_period, choices)
    
    return df
```

```python
forecast_df['start_date'] = pd.to_datetime(forecast_df['start_date'], format ="%Y-%m-%d %H:%M:%S")
```

```python
forecast_processed = extract_from_date(forecast_df)
```

```python
l_object = forecast_processed[["StartStation Name", "weather_main", "season", "day_type", "day_period","day_of_week"]]
```

```python
codes = ohe.transform(l_object).toarray()
feature_names = ohe.get_feature_names(["StartStation Name", "weather_main", "season", "day_type", "day_period","day_of_week"])
```

```python
forecast_processed = pd.concat([forecast_processed[['start_date', 'temp', 'feels_like', 'temp_min', 'temp_max', 'humidity', 'wind_speed', 'year', 'month', 'day', 'hour']].reset_index(drop=True), 
               pd.DataFrame(codes,columns=feature_names).reset_index(drop=True)], axis=1)
```

```python
forecast_processed
```

```python
forecast_processed['hour_sin'] = np.sin(2 * np.pi * forecast_processed['hour'] / 23)
forecast_processed['hour_cos'] = np.cos(2 * np.pi * forecast_processed['hour'] / 23)

forecast_processed['month_sin'] = np.sin(2 * np.pi * forecast_processed['month'] / 12)
forecast_processed['month_cos'] = np.cos(2 * np.pi * forecast_processed['month'] / 12)
```

```python
forecast_processed = forecast_processed.drop(columns=['start_date', 'year', 'month', 'day', 'hour'])
```

```python
forecast_processed[["feels_like", "humidity", "temp", "temp_max", "temp_min", "wind_speed"]] = scaler.transform(forecast_processed[["feels_like", "humidity", "temp", "temp_max", "temp_min", "wind_speed"]])
```
